Types of argument passing 

```python
# 1. Required argument - It is a type a argument where the no. of arguments should be equal to no. of parameters

def func(a,b,c):
    print(f'a:{a}, b:{b}, c:{c} ')

func(10,20,30)
```

```python
# 2. Position Argument - Based on position, value is passed

def func(a,b,c):
    print(f'a:{a}, b:{b}, c:{c} ')

func(20,10,30)
```

```python
# 3. Keyword argument - We specify which arguments value is going to store in which varibale

def func(a,b,c):
    print(f'a:{a}, b:{b}, c:{c} ')

func(b=20,a=10,c=30)
```

```python
# 4. Default arguments - It takes a value by default during function declaration

def func(a,b,c=50):
    print(f'a:{a}, b:{b}, c:{c} ')

func(b=20,a=10)
func(b=20,a=10,c=8)
```

```python
# 5. Varibale length argument(*args) - It run based on the context, we can insert any number of value in this, it accepts the value in form of tuple

def func(*info):
    print(info)

func("Navyajain",8,28,'tjain')
```

```python
# 6. keyword Varibale length argument(**kwargs) - It accepts the data in form of key-value, it accepts the value in form of dictionary

def func(**info):
    print(info)

func(name='Navya',email='navyajain@123')
```

DECORATORS

```python
# Old way to use a decorator

def outer_func(z):
    def inner_func():
        print('Inner hey',z().upper())
        
    return inner_func


def test():
    return('navya jain')

test=outer_func(test)
test()
```

```python
#New way to use a decorator

def outer_func(z):
    def inner_func():
        print('Inner hey',z().upper())

    return inner_func

@outer_func
def test():
    return('navya jain')

test()
```

```python
# Write a Python function to find the maximum of three numbers.

def max(a,b,c):
    if a>b:
        if a>c:
            return(f'{a} is maximum')
    elif b>c:
        return(f'{b} is maximum')
    else:
        return(f'{c} is maximum')
    
max(100,23,27)
```

```python
def square_result(func):
    def answer():
        a=func()
        return a**2
    
    return answer
    


@square_result
def get_num():
    return 5

get_num()
```

```python
def repeat(num):
    def decorate(func):
        def a():
            for i in range(num):
                func()
        return a
    return decorate

@repeat(4)
def greet():
    print("Hi!")

greet()
```

```python
def uppercase(func):
    def a():
        b=func()
        return b.upper()

    return a

def exclamation(func):
    def a():
        b=func()
        return b+'!'

    return a 

@uppercase
@exclamation
def greet():
    return "hello"

greet()

```
