```python
# Write a program to print all Prime numbers between 1 to n.

n = 10
for i in range(1,n+1):
  count = 0
  for j in range(2,i+1):
    if(j<i):
      if i%j==0:
         count+=1
         break
      else:
        j+=1
    else:
      break

  if count == 0:
    print(i,end=' ')
```

```python
# To check if a number is perfect number or not

n=int(input('Enter number: '))
sum=0
for i in range(1,n):
    if n%i==0:
        sum+=i
if sum==n:
    print('Perfect number')
else:
    print('Not a perfect number')
```

```python
# Prime numbers between 1 to n

n=10
prime=[]
i=2
while(i<=n):
    count=0
    for j in range(2,i):
        if i%j==0:
            count+=1
            break
    if count==0:
        prime.append(i)
    i+=1
        

print(prime)
```

```python
# Write a program to check whether a number is Strong number or not

num=145
num2=145
sum=0

while(num>0):
    product=1
    last_digit=num%10
    for i in range(1,last_digit+1):
        product*=i

    sum+=product   
    num//=10

if sum==num2:
    print('Strong number')
else:
    print('Not a strong number')
```

```python
# Write a program to find ones complement of a binary number
# One's complement of a binary number is defined as value
# obtained by inverting all binary bits. It is the result of swapping all
# 1s to 0s and all 0s to 1s.

binary=input('Enter a binary number: ')
ones_complement=''
for bit in binary:
    if bit=='0':
        ones_complement+='1'
    else:
        ones_complement+='0'
print(ones_complement)
```
