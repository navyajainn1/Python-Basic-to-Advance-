```python
# Q 10. Calculating BMI

weight = eval(input("Enter your weight(in kg): "))
heightft = eval(input("Enter your height(in ft.): "))

heightms = heightft/3.28084   # Converting height from feet to meters

bmi = weight / (heightms ** 2)

if(bmi < 18.5):
  print("Underweight")
elif(bmi >= 18.5 and bmi < 24.9):
  print("Normal weight")
elif(bmi >= 24.9 and bmi < 29.9):
  print("overweight")
else:
  print("Obesity")

```

```python
# Q 13. ATM

balance = 100000

option = input("Enter the name of task you want to perform(check balance, deposit money, withdraw money): ")

if(option == "check balance"):

  print(balance)

elif(option == 'deposit money'):

  amt = int(input("Enter the amount of money you want to deposit: "))
  balance+=amt
  print("The total amount of money in your account after depositing money is: ",balance)

elif(option == 'withdraw money'):

  amt = int(input("Enter the amount of money you want to withdraw: "))

  if(amt>balance):
    print("You don't have sufficient amount of money to withdraw")
  else:
    balance-=amt
    print("The total amount of money in your account after withdrawing of money is: ",balance)

else:
  print("Invalid option")
```
