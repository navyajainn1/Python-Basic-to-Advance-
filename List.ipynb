```python
# (LEETCODE) Stock purchase and sell maximum profit

prices = [7,1,5,3,6,4]
profit = 0
for i in range(len(prices)):
    curr  = prices[i]
    for j in range(i+1,len(prices)):
        next = prices[j]
        if next-curr > profit:
            profit = next-curr

print(profit)
```

```python
# Write a Python program to flatten a nested list (list within lists) into a single list.

nested_list = [1, [2, 3], [4, [5, 6], 7], 8]
flattend_list = []

for i in nested_list:
    if type(i) == list:
        for j in i:
            if type(j) == list:
                for k in j:
                    flattend_list.append(k)
            else:
                flattend_list.append(j)
    else:
        flattend_list.append(i)
print(flattend_list)
```

```python
# Write a Python program to find the sum of the elements in a list, excluding the largest
# and smallest element. Don’t use max or min functions

li =  [1, 2, 3, 4, 5] 
n_max =li[0]
n_min = li[0]
total = 0
for i in range(len(li)):
    temp = li[i]
    total+=temp
    if temp<n_min:
        n_min = temp
    if temp>n_max:
        n_max = temp

result = total - n_min - n_max

print('Sum after excluding smallest and largest:',result)

```

```python
# Find the First Non-Repeating Element 
# Given a list of integers, find the first element that appears only once. 

li=[4, 5, 1, 2, 0, 4, 5, 2]
i = 0
while(i<len(li)-1):
    count=0
    j=0
    while(j<len(li)-1):
        if j!=i:
            if li[i]==li[j]:
                count+=1
        j+=1
        
    if count==0:
        print('First element that appears only once:',li[i])
        break
    else:
        i+=1
            
```
