```python
# Hollow traingle

for i in range(5):
    for j in range(4,0,-1):
        if i==4 or j==i:
            print("*",end='')
        else:
            print(" ",end='')
    for j in range(2,5):
        if j==i or i==4:
            print("*",end='')
        else:
            print(" ",end='')
    print()
```

```python
# Pascal Traingle

newlist = []
for i in range(5): #rows
    temp=[]
    for s in range(0,5-i-1):  #spaces
        print(' ',end='')
    for k in range(i+1):    #columns
        if(k==0 or k==i):
            print('1',end=' ')
            temp.append(1)
        else:
            val = newlist[i-1][k] + newlist[i-1][k-1] 
            print(val,end=' ')
            temp.append(val)
    newlist.append(temp)
    print()
```
