File Modes

```python
# 'r' (read): Opens for reading.
# 'w' (write): Opens for writing. Creates the file if it doesn't exist, or truncates (clears) it if it does. 
# 'a' (append): Opens for appending. Creates the file if it doesn't exist, or adds data to the end if it does. 
# 'x' (exclusive creation): Creates a new file.
# 'b' (binary): Used with other modes (e.g., 'rb', 'wb') for binary files like images or videos.
# 't' (text): Default mode for text files (e.g., 'rt', 'wt').
```

```python
# To open file:

# If our both files is at same location we can then just write the name of the file we want to use and 
# if the location is difference then we have to give the fulll path of file

f=open('data.txt') 

# To read file

data=f.read()
print(data)

# To close file
f.close()
```

```python
# To write in a file

f=open('data.txt','w')  # w is a access mode for write( by default its r(read) mode)
f.write('Navya jain')
f.close() 
```

```python
f=open('data.txt','r+')  # r+ is a access mode for read and write both
out=f.read() 
f.write('Data science')
f.close() 
print(out)
```

```python
# New syntax - Here no need to close the file

with open('data.txt','r+') as f: 
    out=f.read() 
    f.write('Data science')

print(out)
```

```python
with open('data.txt','r+') as f: 
    print(f.tell()) # It tells the position of cursor
    out=f.read()
    print(f.tell()) 
    f.seek(2)      # It seeks the position according to us
    print(f.tell())

print(out)
```

```python
with open('data.txt') as f:
    print(f.readline().strip()) #it read one line at a time
    print(f.readline().strip())
```

```python
with open('data.txt') as f:
    print(f.readlines()) #it read all lines in a list
```

```python
# We can also use loop instead of readlines

with open('data.txt') as f:
    for line in f:
      print(line) 
```

```python
with open('data.txt') as f:
    for line in f:
      print(line.strip()) # Strip to remove /n 
```

```python
data =[['Name','Age'],['Navya',19],['Tanishq',19]]
import csv
with open('info.txt','w',newline='') as csvfile:
    csv_writer=csv.writer(csvfile)
    csv_writer.writerow(data[0])
    csv_writer.writerows(data[1:])

print('success')
    
```

```python
num=2
for i in range(1,11):
    with open('data.txt','a',newline='') as file:
        file.write(str(num*i))
        file.write('\n')
    
```

```python
# Making prime and non prime numbers different files

for i in range(4,41):
    count=0
    for j in range(2,i):
        if i%j==0:
            count+=1
            break
    if count>0:
         with open('prime.txt','a',newline='') as file:
            file.write(str(i))
            file.write('\n')
    else:
        with open('non_prime.txt','a',newline='') as file:
            file.write(str(i))
            file.write('\n')

         
```
