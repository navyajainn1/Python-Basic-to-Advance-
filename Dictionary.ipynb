```python
# Check the occurence of element in a list and insert in dictionary

li = [10,20,30,10,40,50]
mydict = {}
for i in range(len(li)):
    count = 1
    temp = li[i]
    if temp not in mydict:
        mydict[temp] = count
    else:
        mydict[temp] = count+1

print(mydict)
```

```python
# count the occurence of vowels 

s = 'aeibujiaa'
d={}
for i in s:
    if i in'aeiou':
        if i not in d:
            d[i]=1
        else:
            d[i]=d[i]+1

print(d)
```

```python
# Two sum problem

a=[3,5,7,11,16,27]
target = 21
mydictionary={}
for i in range(len(a)):
    data=a[i]   
    if data not in mydictionary:
        mydictionary[data]=i
print(mydictionary)

for x in mydictionary:
    val = mydictionary[x]
    if (target-x in mydictionary):
        print('We found value:',mydictionary[target-x],val)
```
